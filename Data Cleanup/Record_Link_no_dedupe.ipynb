```python
import pandas as pd
import py_stringmatching as sm
```

```python
sunshine = pd.read_csv('../Data/Outputs_Cleanup/physicians_info_dedup.csv',dtype=str)
```

```python
part_d = pd.read_csv('../Data/Outputs_Cleanup/prescriber_dedup.csv',dtype=str)
```

```python
part_d.head()
```

```python
sunshine.head()
```

```python
def processType(x):
    s = x.split('|')
    return s[-1]
```

```python
def processData(df):
    cols = df.columns.to_list()
    for col in cols:
        df[col] = df[col].astype(str)
        df[col] = df[col].apply(lambda x: x.lower().strip())
    df['type'] = df['type'].apply(lambda x: processType(x))
    df['combined'] = df['fname'] + ' ' + df['lname'] + ' ' + df['city'] + ' ' + df['type']
    return df
```

```python
sunshine = processData(sunshine)
sunshine.head()
```

```python
part_d = processData(part_d)
part_d.head()
```

```python
part_d_dups = part_d.drop(part_d[part_d.duplicated(subset='combined',keep=False)==False].index)
sunshine_dups = sunshine.drop(sunshine[sunshine.duplicated(subset='combined',keep=False)==False].index)
```

```python
part_d_dups.to_csv('../Data/Outputs_Cleanup/part_d_dups.csv',index=False)
sunshine_dups.to_csv('../Data/Outputs_Cleanup/sunshine_dups.csv',index=False)
```

```python
part_d.drop(index=part_d_dups.index,inplace=True)
sunshine.drop(index=sunshine_dups.index,inplace=True)
```

Here we will do a naive matching. Based on just the 'combined' column. If there is an exact match we will accept this.
